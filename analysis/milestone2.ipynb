The likeliness to vote by partie is evenly split

### 2. What about undecided voters?

```python
df8 = df[(df["Q23"]==3)]
print('Number of undecided voters: {0}'.format(len(df8)))
df12 = df[(df["Q23"] == 3) & (df["Q5"]==2)]
print('Number of undecided voters who beleive the winner of the election will not make a difference in the state of the country: {0}'.format(len(df12)))
df9 = df[(df["Q23"] == 3) & (df["Q22"] == 6)]
print('Number of undecided voters who beleive their vote do not matter: {0}'.format(len(df9)))
df10 = df[(df["Q23"] == 3) & (df["Q22"] == 2)]
print('Number of undecided voters who beleive the political system does not serve their needs: {0}'.format(len(df10)))
df11 = df[(df["Q20"] == 2) & (df["Q21"] == 1)]
print('Number of voters who plan to vote but had not registered at the time of the survey: {0}'.format(len(df11)))
```

Interesting findings from undecided voters analysis

### 3. Non-voters

```python
df13 = df[(df["Q21"]==2)]
print('Number of people who will not vote: {0}'.format(len(df13)))
df14 = df[(df["Q21"]==3)]
print('Number of people who do not know if they will vote: {0}'.format(len(df14)))
df15 = df[(df["Q21"] == 2) | (df["Q21"] == 3) & (df["Q5"]==2)]
print('Number of voters who will not votes or are not sure if they will vote and who beleive the winner of the election will not make a difference in the state of the country: {0}'.format(len(df15)))
df16 = df[(df["Q21"] == 2) | (df["Q21"] == 3) & (df["Q22"] == 6)]
print('Number of voters who will not votes or are not sure if they will vote and who feel their vote does not matter: {0}'.format(len(df16)))
df17 = df[(df["Q21"] == 2) | (df["Q21"] == 3) & (df["Q22"] == 2)]
print('Number of voters who will not votes or are not sure if they will vote and who beleive the political system does not serve their needs: {0}'.format(len(df17)))
```

Interesting finds from non-voters and potential non-voters

For the three following charts
    1 Represents a yes
    2 Represents a no
    
    -1 indicates that a participant did not answer this question

```python
df18 = df[(df["Q21"] == 2) | (df["Q21"] == 3)]

boxplot = sns.countplot(y ='Q5', data = df18)
boxplot.set_title('Do non-voters think the results of the election will make a difference?')
boxplot.set_ylabel('Opinion')
boxplot.set_xlabel('Number of participants who are non-voters')
plt.show()

boxplot = sns.countplot(y ='Q5', data = df18)
boxplot.set_title('Do non-voters think their vote matters?')
boxplot.set_ylabel('Opinion')
boxplot.set_xlabel('Number of participants who are non-voters')
plt.show()

boxplot = sns.countplot(y ='Q5', data = df18)
boxplot.set_title('Do non-voters think elected officials have their best interest in mind?')
boxplot.set_ylabel('Opinion')
boxplot.set_xlabel('Number of participants who are non-voters')
plt.show()
```

### 4. How US voters vote?

For the following chart: 
 1 = Voting by mail
 2 = Early voting
 3 = In person on election day 
 4 = other
 
-1 indicates that a participant did not answer this question.

```python
boxplot = sns.countplot(y ='Q24', data = df)
boxplot.set_title('How voters planned to vote in the 2020 election')
boxplot.set_ylabel('Voting options')
boxplot.set_xlabel('Number of participants')
plt.show()

df6 = df[(df["Q33"] == 2)]
boxplot = sns.countplot(y ='Q24', data = df6)
boxplot.set_title('How democrates planned to vote in the 2020 election')
boxplot.set_ylabel('Voting options')
boxplot.set_xlabel('Number of participants')
plt.show()

df7 = df[(df["Q33"] == 1)]
boxplot = sns.countplot(y ='Q24', data = df7)
boxplot.set_title('How republicans planned to vote in the 2020 election')
boxplot.set_ylabel('Voting options')
boxplot.set_xlabel('Number of participants')
plt.show()
```

The ratio between mailing voting (option 1) and in person voting (option 3) seems pretty even for the overall population. However, democrates are more likely to vote by mail and republicans are more likely to vote in person. 

## Conclusion

Belonging to a certain party does not make someone less or more likely to vote however, it will influence how a voter chooses to vote. Voters who are undecided are more likely to think that the results of the election will not make a difference but most of them still beleive that their vote matters and that election officials have their best interests in mind. On the contrary, non-voters or people who are unsure if they will vote beleive that the elected president of the 2020 election will not make a difference for the country but they are also very likely to beleive that their vote does not matter and that elected officials do not have their best interests in mind.
